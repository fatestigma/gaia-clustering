Add `describe` method to gastrodon's `RemoteEndpoint`.

````python
from gastrodon import _parseQuery
from SPARQLWrapper import SPARQLWrapper, N3
from rdflib import Graph
def describe(self, sparql:str):
    return self._describe(sparql).serialize(format='n3').decode()

def _describe(self, sparql:str):
    that = endpoint._wrapper()
    that.setQuery(endpoint._prepend_namespaces(sparql, _parseQuery))
    that.setReturnFormat(N3)
    results = that.query().convert()
    g = Graph()
    g.parse(data=results, format="n3")
    return g

RemoteEndpoint.describe = describe
RemoteEndpoint._describe = _describe
````

## Extract entities

````python
def to_int(s):
    return int(s) if isinstance(s, str) or isinstance(s, int) else -1
````

````python
rpi_entity_with_justification = endpoint.select("""
SELECT ?e ?type ?label ?target ?source ?start ?end ?justificationType {
    ?e a aida:Entity ;
       aida:system <http://www.rpi.edu> .
    ?statement a rdf:Statement ;
               rdf:subject ?e ;
               rdf:predicate rdf:type ;
               rdf:object ?type ;
               aida:justifiedBy ?justification .
    OPTIONAL { ?e aida:link/aida:linkTarget ?target }
    OPTIONAL { ?justification skos:prefLabel ?label }
    OPTIONAL { ?justification aida:source ?source }
    OPTIONAL { ?justification aida:startOffset ?start }
    OPTIONAL { ?justification aida:endOffsetInclusive ?end }
    OPTIONAL { ?justification aida:privateData ?typePrivate .
               ?typePrivate aida:system <http://www.rpi.edu> ;
                            aida:jsonContent ?justificationType }
}
""")
# convert start and end to int, -1 if it's None
rpi_entity_with_justification.start = rpi_entity_with_justification.start.apply(to_int)
rpi_entity_with_justification.end = rpi_entity_with_justification.end.apply(to_int)
rpi_entity_with_justification.justificationType = rpi_entity_with_justification.justificationType.apply(lambda s: json.loads(s).get('justificationType'))
rpi_entity_with_justification
````

````python
rpi_event_with_justification = endpoint.select("""
SELECT DISTINCT ?e ?type ?label ?source ?start ?end ?justificationType {
    ?e a aida:Event ;
       aida:system <http://www.rpi.edu> .
    ?statement a rdf:Statement ;
               rdf:subject ?e ;
               rdf:predicate rdf:type ;
               rdf:object ?type ;
               aida:justifiedBy ?justification .
    OPTIONAL { ?justification skos:prefLabel ?label }
    OPTIONAL { ?justification aida:source ?source }
    OPTIONAL { ?justification aida:startOffset ?start }
    OPTIONAL { ?justification aida:endOffsetInclusive ?end }
    OPTIONAL { ?justification aida:privateData ?typePrivate .
               ?typePrivate aida:system <http://www.rpi.edu> ;
                            aida:jsonContent ?justificationType }
}
""")
rpi_event_with_justification.start = rpi_event_with_justification.start.apply(to_int)
rpi_event_with_justification.end = rpi_event_with_justification.end.apply(to_int)
rpi_event_with_justification.justificationType = rpi_event_with_justification.justificationType.apply(lambda s: json.loads(s).get('justificationType'))
rpi_event_with_justification
````

## Extract event roles

````python
print(endpoint.describe("""
describe ?statement {
    ?e a aida:Event ;
       aida:system <http://www.rpi.edu> .
    ?statement a rdf:Statement ;
               rdf:subject ?e ;
               rdf:predicate ?p ;
               rdf:object ?o 
    FILTER (?p != rdf:type)
}
limit 1
"""))
````

````python
rpi_event_roles = endpoint.select("""
SELECT DISTINCT ?e ?p ?o {
    ?e a aida:Event ;
       aida:system <http://www.rpi.edu> .
    ?statement a rdf:Statement ;
               rdf:subject ?e ;
               rdf:predicate ?p ;
               rdf:object ?o ;
               aida:system <http://www.rpi.edu> .
    FILTER (?p != rdf:type)
}
""")
rpi_event_roles
````

````python
rpi_relation_with_justification = endpoint.select("""
SELECT DISTINCT ?e ?type ?source ?start ?end {
    ?e a aida:Relation ;
       aida:system <http://www.rpi.edu> .
    ?statement a rdf:Statement ;
               rdf:subject ?e ;
               rdf:predicate rdf:type ;
               rdf:object ?type ;
               aida:justifiedBy ?justification 
    OPTIONAL { ?justification aida:source ?source }
    OPTIONAL { ?justification aida:startOffset ?start }
    OPTIONAL { ?justification aida:endOffsetInclusive ?end }
}
""")
rpi_relation_with_justification.start = rpi_relation_with_justification.start.apply(to_int)
rpi_relation_with_justification.end = rpi_relation_with_justification.end.apply(to_int)
rpi_relation_with_justification
````

## Document languages

````python
document_types = endpoint.select("""
SELECT DISTINCT ?source ?fileType {
    ?justification a aida:TextJustification ;
                   aida:system <http://www.rpi.edu> ;
                   aida:source ?source ;
                   aida:privateData ?filePrivate .
    ?filePrivate aida:system <http://www.rpi.edu/fileType> ;
                 aida:jsonContent ?fileType
}
""")
document_types['lang'] = document_types.fileType.apply(lambda s: json.loads(s).get('fileType'))
document_types
````
